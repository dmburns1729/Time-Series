```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import os
# set random seed
SEED = 321
np.random.seed(SEED)

#import tick customization tools
import matplotlib.ticker as mticks
import matplotlib.dates as mdates

pd.set_option('display.float_format',lambda x:f"{x:,.2f}")

sns.set_context('notebook',font_scale=0.9)
plt.style.use(['ggplot'])
```

```python
df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vQN3Ksa9szQuO4G1-msXWAp17KtVHQCBnuEieu_auu1wSiBf3-krHusIx5VBMkihxj-KZLBosDIGEyR/pub?output=csv")
```

```python
display(df.head())
df.info
```

```python
df_copy = df.copy()
```

```python
df_filtered = df_copy.head(4)
```

```python
df_filtered
```

```python
melted = pd.melt(df_filtered.drop(columns=['RegionID','SizeRank','RegionName','RegionType','StateName', 'State','CountyName']),
                 id_vars=['Metro'],
                value_name='Home Price',var_name="Date")
melted
```

```python
melted['Date'] = pd.to_datetime(melted['Date'])
melted.info()
```

```python
melted = melted.set_index("Date")
melted
```

```python
df_ts = melted.groupby('Metro').resample("M").mean()
df_ts
```

### This workaround isn't working
NOTE: Pandas' .plot method currently has a bug with 4-digit year axis labels. The tick labels misbehave when being formatted with matplotlib and will produce a mostly-blank axis with 1 year labeled, like the screenshot below:

To get around this issue, first create your fig and ax using plt.subplots() Now you may use pandas .plot method to plot onto this figure by adding "ax=ax" to .plot. fig, ax = plt.subplots(figsize=(12,3)) df.plot(ax=ax)

```python
def thousands(x,pos):
    """function for use wth matplotlib FuncFormatter -  formats money in millions"""
    new_x = x/1_000
    return f"${new_x:,.0f}K"
# Create the formatters
fmt_thous = mticks.FuncFormatter(thousands)
```

```python
plt.rcParams['figure.figsize'] = (12,5)
```

```python
ax = unstacked['Home Price'].plot()
ax.set(title='Home Prices by City', ylabel='Home Price', xlabel="Date")

# customize major ticks 
major_loc =  mdates.YearLocator()
major_fmt = mdates.DateFormatter("%Y")


str_fmt = mticks.StrMethodFormatter("${x:,.0f}")
ax.yaxis.set_major_formatter(fmt_thous)

```
